**Missing values**

```python
# here I am going to check the missing values in the whole data set through each column.

for col in df.columns:
  check_pct = np.mean(df[col].isnull())
  print(" {} - {:<2}%".format(col, check_pct))
```

```python
"""
df: This refers to a DataFrame, which is a 2-dimensional labeled data structure with columns of potentially different types.
 It's similar to a spreadsheet or SQL table, or a dictionary of Series objects.

.dropna(): This is a method of a DataFrame that allows you to drop rows with missing values (NaN).

subset=['budget', 'gross']: The subset parameter specifies which columns to consider when checking for missing values.
 In this case, only rows where either 'budget' or 'gross' have missing values will be considered for removal.

inplace=True: The inplace parameter, when set to True, modifies the original DataFrame directly without returning a new one. If inplace were set to False (or not provided, as its default value is False), the method would return a new DataFrame with the rows dropped, and the original DataFrame would remain unchanged."""

# So as we see here some important data that needs to clean so for that we remove the rows of "rating", "Buget" and "Gross" missing value.

df.dropna(subset=['budget', 'gross','rating'], inplace=True)
for col1 in df.columns:
  check_pct1 = np.mean(df[col1].isnull())
  print(" {} - {:<2}%".format(col1, check_pct1))
```

df['released'].astype(str): This converts the 'released' column of the DataFrame df to strings. This ensures that the subsequent string operations can be applied without issues, even if the original column had mixed types or wasn't a string type.

.str.extract(r'(\d{4})'): This extracts a sequence of four digits from each string in the 'released' column. Specifically:

.str: This allows you to apply string methods to a Series.
.extract(): This method extracts a specific pattern from a string.
r'(\d{4})': This is a regular expression. Here's a breakdown of the expression:
\d: Matches any digit (equivalent to [0-9]).
{4}: Specifies that the previous pattern (\d) should be matched exactly 4 times.
() (parentheses): This creates a capturing group, which means that only the sequence of four digits will be extracted and returned.
df['yearcorrected'] = ...: This assigns the extracted year (the sequence of four digits) to a new column in the DataFrame named 'yearcorrected'.

In summary, this line of code extracts a four-digit year from the 'released' column of the DataFrame df and stores it in a new column named 'yearcorrected'. If a row in the 'released' column doesn't have a four-digit sequence, the corresponding entry in the 'yearcorrected' column will be NaN.

```python
# Now chaning the year column becuase the release date and year column is different.
df['yearcorrected'] = df['released'].astype(str).str.extract(r'(\d{4})')
df.head()
```

#Data Analysis

**Highest Correlations to Gross Earnings**

In this section the analysis is focused on finding the highest correlations to gross earnings. The following chunk of code converts categorical columns in the DataFrame 'df' into numerical values using label encoding and stores the result in the new DataFrame 'df_numerized':

```python
df_numerized = copy.deepcopy(df)
```

```python
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

After the 'for' loop completes, the DataFrame 'df_numerized' contains the same data as the original DataFrame 'df', but the categorical columns have been replaced with corresponding integer codes. This transformation is useful when you want to use machine learning algorithms that require numerical inputs and cannot directly handle categorical data. By doing this, it will be possible to see all the correlations in a heatmap visualization. The method used to find the correlations between variables is the Pearson method:

```python
correlation_matrix = df_numerized.corr(method='pearson')
#correlation_matrix = df_numerized.corr(method='kendall')
#correlation_matrix = df_numerized.corr(method='spearman')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Metric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

The bigger the coefficient the higher the correlation. So, from the heatmap 'budget' and 'votes' have the biggest coefficient, 0.74 and 0.61, respectively. In this case, there are so many correlations, so it will be useful to obtain only the highest correlations by doing the following

```python
correlation_mat = df_numerized.corr(method='pearson')

corr_pairs = correlation_mat.unstack()  #correlation matrix from a 2D table format into a long 1D Series

sorted_pairs = corr_pairs.sort_values() #sort_values() method will sort this Series in ascending order based on its values

high_corr = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < 1)]

high_corr
```

* **'budget'** and **'votes'** have the highest correlations with **'gross'** earnings
